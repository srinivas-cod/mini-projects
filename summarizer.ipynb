```python
from transformers import pipeline
summarizer = pipeline("summarization", model = "facebook/bart-large-cnn")
```

```python
def five_w_one_h_keyword(text):
    summary_text = summarizer(text, max_length=3000, min_length=30, do_sample=False)[0]['summary_text']
    five_w_h = {"What": "", "Why": "", "How": "", "When": "", "Where": "", "Who": "", "Short Summary": summary_text}

    sentences = summary_text.split(". ")
    for sentence in sentences:
        lower = sentence.lower()
        if "because" in lower or "used" in lower:
            five_w_h["Why"] = sentence
        elif "process" in lower or "method" in lower:
            five_w_h["How"] = sentence
        elif "time" in lower or "when" in lower:
            five_w_h["When"] = sentence
        elif "place" in lower or "where" in lower:
            five_w_h["Where"] = sentence
        elif "who" in lower or "person" in lower:
            five_w_h["Who"] = sentence
        elif five_w_h["What"] == "":
            five_w_h["What"] = sentence

    return five_w_h
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
```

### Subtask:
Create a set of questions for each of the 5W1H (What, Why, How, When, Where, Who).


**Reasoning**:
Create a dictionary containing lists of questions for each of the 5W1H categories.



```python
questions_5w1h = {
    "What": [
        "What is this about?",
        "What happened?",
        "What is the main topic?",
        "What is the subject?",
        "What was done?"
    ],
    "Why": [
        "Why did this happen?",
        "What is the reason?",
        "Why was this done?",
        "What was the cause?"
    ],
    "How": [
        "How did this happen?",
        "How was this done?",
        "What is the method?",
        "What is the process?"
    ],
    "When": [
        "When did this happen?",
        "What is the date?",
        "What is the time?",
        "When will this take place?"
    ],
    "Where": [
        "Where did this happen?",
        "What is the location?",
        "Where is it located?",
        "What is the venue?"
    ],
    "Who": [
        "Who is involved?",
        "Who did this?",
        "Who are the participants?",
        "Who is the subject?"
    ]
}
```

## Update the extraction function

### Subtask:
Modify the `five_w_one_h_keyword` function to use the question answering model to find the answers to the defined questions within the summarized text.


**Reasoning**:
Modify the existing function to use the question answering model to extract 5W1H information from the summarized text.



```python
def five_w_one_h_keyword(text):
    summary_text = summarizer(text, max_length=3000, min_length=30, do_sample=False)[0]['summary_text']
    five_w_h = {"What": "", "Why": "", "How": "", "When": "", "Where": "", "Who": "", "Short Summary": summary_text}

    for category, questions in questions_5w1h.items():
        best_answer = ""
        best_score = -1.0
        for question in questions:
            qa_input = {
                'question': question,
                'context': summary_text
            }
            answer = qa_pipeline(qa_input)
            if answer['score'] > best_score:
                best_score = answer['score']
                best_answer = answer['answer']
        five_w_h[category] = best_answer

    return five_w_h
```
